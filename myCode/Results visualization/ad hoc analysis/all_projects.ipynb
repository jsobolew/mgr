```python
project = 'rehersal Alexnet MNIST Task IL tr-t split v2'
runs = api.runs(f"qba/{project}")
```

```python
runs_list = []
for run in runs:
    runs_list.append(run.config)
df = pd.DataFrame(runs_list).astype(str)
```

```python
faulty_runs = []

for i in range(len(df)):
    if df.iloc[i].classes_list != 'nan':
        classes_list = json.loads(df.iloc[i].classes_list)
        for j, task in enumerate(classes_list):
            if 0 in task:
                if task[1] == 0:
                    faulty_runs.append(i)
```

```python
df
```

```python
df.loc[faulty_runs, ['batch_size', 'num_classes', 'classes_per_task']] = df.loc[faulty_runs, ['batch_size', 'num_classes', 'classes_per_task']].astype(float).astype(int)
```

```python
setup_columns = ['project',
'setup',
'learning_rate',
'architecture',
'dataset',
'rehearsal_dataset',
'epochs',
'batch_size_rehearsal',
'batch_size',
'pretraining',
'num_classes',
'classes_per_task',
'optimizer']
```

```python
df_faulty_runs = df.iloc[faulty_runs]

df_faulty_runs.loc[faulty_runs, ['batch_size', 'num_classes', 'classes_per_task']] = df_faulty_runs.loc[faulty_runs, ['batch_size', 'num_classes', 'classes_per_task']].astype(float).astype(int)
```

```python
run_params = []
for i in range(len(df_faulty_runs)):
    execution_params = ""
    for setup_column in setup_columns:
        execution_params += setup_column+'="'+str(df_faulty_runs.iloc[i][setup_column])+'"'+" "
    run_params.append(execution_params)
```

```python
f = open(f'{project}.pickle',"wb")
pickle.dump(run_params ,f)
f.close()
```
