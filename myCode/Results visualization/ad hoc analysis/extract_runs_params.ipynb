```python
project = 'rehersal ResNet CIFAR10 Task IL'
runs = api.runs(f"qba/{project}")
```

```python
runs_list = []
for run in runs:
    runs_list.append(run.config)
df = pd.DataFrame(runs_list).astype(str)
```

```python
setup_columns = ['project',
'setup',
'learning_rate',
'architecture',
'dataset',
'rehearsal_dataset',
'epochs',
'batch_size_rehearsal',
'batch_size',
'pretraining',
'num_classes',
'classes_per_task',
'optimizer']
```

```python
df_to_run = df[df['epochs'] == '5']
```

```python
df_to_run = df_to_run[setup_columns].drop_duplicates().sort_values(by=['rehearsal_dataset'])
```

```python
# df_to_run = df.copy()

df_to_run.loc[:,['batch_size', 'batch_size_rehearsal', 'num_classes', 'classes_per_task']] = df_to_run.loc[:,['batch_size', 'batch_size_rehearsal', 'num_classes', 'classes_per_task']].astype(float).astype(int)
```

```python
n_times = 5

run_params = []
for i in range(len(df_to_run)):
    for n in range(n_times):
        execution_params = ""
        for setup_column in setup_columns:
            execution_params += setup_column+'="'+str(df_to_run.iloc[i][setup_column])+'"'+" "
        run_params.append(execution_params)
```

```python
runs_per_file = 5

header = """#!/bin/bash
#SBATCH -A plgdyplomancipw-gpu-a100
#SBATCH -p plgrid-gpu-a100
#SBATCH -t 2880
#SBATCH --ntasks 1
#SBATCH --gres gpu:1
#SBATCH --mem 40G
#SBATCH --nodes 1
#SBATCH -o slurm_out/slurm-%j.log
#SBATCH -J "noise rehersal"

# LOAD MODULES, INSERT CODE, AND RUN YOUR PROGRAMS HERE
#!/bin/bash
module load Miniconda3/4.9.2

#conda activate /net/tscratch/people/plgjsobolewski/conda_env

"""

for i, run_param in enumerate(run_params):
    if i % runs_per_file == 0:
        f = open(f'../../schedule_scripts/schedule_{i//runs_per_file}.sh', "w")
        f.writelines(header)
        wrote_lins = 0

    command = "/net/tscratch/people/plgjsobolewski/conda_env/bin/python ../run_experiment.py "+run_param
    wrote_lins += 1
    if wrote_lins != runs_per_file:
        command += "&"+"\n"

    f.writelines(command)

    if i % runs_per_file == 0 and wrote_lins == runs_per_file:
        f.close()

# meta run
f = open(f'../../schedule_scripts/meta_shedule.sh', "w")
for i in range(len(run_params)//runs_per_file):
    f.writelines(f"sbatch schedule_{i}.sh\n")
f.close()
```
